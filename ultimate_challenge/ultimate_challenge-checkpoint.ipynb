### Data description
- city: city this user signed up in

#### Active date
Because we want to see if the drivers are active by compared their last trip. We will get the most recent trip and for each user, we will take the last day they have a trip then minus the most recent trip from the dataset.


```python
# Calculate most_recent_trip
most_recent_trip = max(users.last_trip_date)
ealiest_signup_date = min(users.signup_date)

print('Most recent trip:', most_recent_trip)
print('Earliest signup date:', ealiest_signup_date)
```

Now, we will observe users who were retained. A user is retained if he/she was active (i.e. took a trip) in the preceding 30 days.

```python
# Create a new columns to save most recent trip
users['most_recent_trip']=pd.to_datetime(most_recent_trip)
users['last_trip_date']=pd.to_datetime(users.last_trip_date)

# Calculate the different between most recent trip and the last trip he/she drived.
users['active_date']=(users.most_recent_trip-users.last_trip_date).dt.days
users['active_date'].head()
```

```python
# Create a retained columns to save active drivers who had recent trips <30 days
users['retained'] = users.active_date.apply(lambda x: 1 if x < 30 else 0)
users[['active_date','retained']].head()
```
