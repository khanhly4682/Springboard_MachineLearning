# Data Wrangling for Zillow's Home Value Prediction

__Objective:__

This project is to predict the log-error between their Zestimate and the actual sale price, given all the features of a home. The dataset consists of 2 separate csv files containing logerror house price prediction and properties of the house.

The log error is defined as:
- _logerror=log(Zestimate)−log(SalePrice)_

and it is recorded in the transactions file train.csv.

https://www.kaggle.com/c/zillow-prize-1/data?

- train_2017.csv

#### 1.3. Correlation Analysis

Now, let's check the correlationship between target variable and other features and the relationship between features themselves. Features with high correlation are more linearly dependent and hence have almost the same effect on the dependent variable. So when two features have high correlation, we can drop one of the two features.

```python
# Check feature correlation with target 'logerror'
corr = new_df[numerical_cols].corr()

plt.figure(figsize=(16,14))
sns.heatmap(corr, annot=True)
plt.title('Feature correlation', fontsize=16)
plt.show()
```

#### Comments:
The target value '_logerror_' is not highly correlated with other features. However, we can see high correlations between other features:
- _bedroomcnt_ is highly correlated with _bathroomcnt_
- _roomcnt_ is highly correlated with _fips_
- _fullbathcnt_ is highly correlated with _bathroomcnt_
etc.

So we can consider to remove the highly corelated features in the dataset such as _bathroomcnt_, and _fullbathcnt_ because both features are highly correlated with _bedroomcnt_. We also drop _roomcnt_ because it is correlated with _fips_ and the number of bedrooms (_bedroomcnt_) is more also similar count or it can be more significant than the number of room (_roomcnt_).

So we will delete features with correlation larger than 0.9.

```python
# Delete highly correlate features with threshold > 0.9
corr = new_df[numerical_cols].corr()

def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                #if colname in dataset.columns:
                #    del dataset[colname] # deleting the column from the dataset

    #return dataset.columns.values.tolist()
    return list(col_corr)

corr_cols = correlation(new_df, 0.9)

# Drop highly correclated features
new_df = new_df.drop(columns=corr_cols,axis=1)

print(corr_cols)
print(new_df.shape)
```

We have highly correlated features that are ['finishedsquarefeet12', 'rawcensustractandblock', 'landtaxvaluedollarcnt', 'censustractandblock', 'taxamount', 'fullbathcnt', 'calculatedbathnbr']. After removing those features from the data, we have 23 columns left.

## Data Imputation
For numerical features that contain id, we fill the missing values with the most common id code, including FIPs codes.

```python
# Fill missing ID with the most frequent value.
id_cols = [x for x in new_df.columns if 'id' in x]
id_cols.append('fips')
for col in id_cols:
    new_df[col]=new_df[col].fillna(new_df[col].mode()[0])
```

For the other numerical features, we fill the missing values with the mean value.

```python
# Select numerical cols that are not id cols.
categorical_cols = new_df.select_dtypes(['object']).columns.values.tolist()
numerical_cols = new_df.select_dtypes(['int','float']).columns.values.tolist()
numerical_cols_not_id = [x for x in new_df.columns if x not in id_cols and x not in categorical_cols]
#print(numerical_cols2)

for col in numerical_cols_not_id :
    new_df[col]=new_df[col].fillna(new_df[col].mean())
```

## Categorical Data Imputation

```python
# Select categorical columns
categorical_cols = new_df.select_dtypes(['object']).columns.values.tolist()
#categorical_cols

# Fill missing columns with the most frequent value.
for col in categorical_cols:
    new_df[col]=new_df[col].fillna(new_df[col].mode()[0])
```
